```python
import pandas as pd
import matplotlib.pyplot as plt

df_last = pd.read_csv("data/플젝/구별로 나눈 csv/중랑구.csv", encoding="euc-kr")
df_last.shape   #행 열 순서로 나옴

plt.rc("font", family="Malgun Gothic")

df_last = pd.read_csv("data/플젝/구별로 나눈 csv/중랑구.csv", encoding="euc-kr")
df_last2 = pd.read_csv("data/플젝/구별로 나눈 csv/중랑구.csv", encoding="euc-kr")

pd.pivot_table(df_last, index=[" 일자"], values=[" 단위면적(㎡)당 매매가(만원)","단위면적(㎡)당 전세가(만원)"], aggfunc="mean")
```

```python
pd.pivot_table(df_last2, index=[" 일자"], values=[" 출생자 수"], aggfunc="mean")
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl

plt.rc("font", family="Malgun Gothic")
plt.rc('font', size=40)        # 기본 폰트 크기
plt.rc('axes', labelsize=20)   # x,y축 label 폰트 크기
plt.rc('xtick', labelsize=15)  # x축 눈금 폰트 크기 
plt.rc('ytick', labelsize=20)  # y축 눈금 폰트 크기
plt.rc('legend', fontsize=20)  # 범례 폰트 크기
plt.rc('figure', titlesize=50) # figure title 폰트 크기


g = df_last.groupby([" 일자"])[" 단위면적(㎡)당 매매가(만원)"].mean()
g2 = df_last.groupby([" 일자"])["단위면적(㎡)당 전세가(만원)"].mean()
g3 = df_last2.groupby([" 일자"])[" 출생자 수"].mean()


fig, ax1 = plt.subplots(figsize=(20, 15))
ax1.plot(g,label="단위면적(㎡)당 매매가(만원)")
ax1.plot(g2,label="단위면적(㎡)당 전세가(만원)")

plt.xticks(rotation=90)

plt.legend(loc="upper right")


plt.xlabel('일자')
plt.ylabel('단위면적(㎡)당 금액(만원)')


ax2 = ax1.twinx()
plt.ylabel('출생자 수')

ax2.plot(g3, color='deeppink',label="출생자 수")

plt.title('중랑구')

plt.legend(loc="upper left")

plt.show()
```
