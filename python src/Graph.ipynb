```python
import matplotlib.pyplot as plt
plt.rc("font", family="Malgun Gothic")
```

```python
g = df_last.groupby([" 년도"])[" 단위면적(㎡)당 매매가(만원)"].mean()
g.plot()
```

```python
g2 = df_last.groupby([" 년도"])["단위면적(㎡)당 전세가(만원)"].mean()
g2.plot()
```

```python
g3 = df_last.groupby([" 년도"])[" 혼인 건수"].mean()
g3.plot()
```

```python
g4 = df_last.groupby([" 년도"])[" 출생자 수"].mean()
g4.plot()
```

```python
g.plot()
g2.plot()
g4.plot()

fig, ax1 = plt.subplots()
ax1.plot()

plt.xlabel('년도')
plt.ylabel('단위면적(㎡)당 전세가(만원)')

plt.ylabel('단위면적(㎡)당 전세가(만원)')
plt.legend()
plt.title('강남구')
plt.show()
```
