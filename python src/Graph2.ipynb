```python
import pandas as pd
df_last = pd.read_csv("data/플젝/test2.csv", encoding="euc-kr")
df_last2 = pd.read_csv("data/플젝/test2.csv", encoding="euc-kr")

pd.pivot_table(df_last, index=[" 년도"], values=[" 단위면적(㎡)당 매매가(만원)","단위면적(㎡)당 전세가(만원)"], aggfunc="mean")
```

```python
pd.pivot_table(df_last2, index=[" 년도"], values=[" 출생자 수"], aggfunc="mean")
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl

plt.rc("font", family="Malgun Gothic")
plt.rc('font', size=40)        # 기본 폰트 크기
plt.rc('axes', labelsize=20)   # x,y축 label 폰트 크기
plt.rc('xtick', labelsize=15)  # x축 눈금 폰트 크기 
plt.rc('ytick', labelsize=20)  # y축 눈금 폰트 크기
plt.rc('legend', fontsize=20)  # 범례 폰트 크기
plt.rc('figure', titlesize=50) # figure title 폰트 크기
```

```python

g = df_last.groupby([" 년도"])[" 단위면적(㎡)당 매매가(만원)"].mean()
g2 = df_last.groupby([" 년도"])["단위면적(㎡)당 전세가(만원)"].mean()
g3 = df_last2.groupby([" 년도"])[" 출생자 수"].mean()
```

```python

fig, ax1 = plt.subplots(figsize=(20, 15))
ax1.plot(g,label="단위면적(㎡)당 매매가(만원)")
ax1.plot(g2,label="단위면적(㎡)당 전세가(만원)")

plt.xticks(rotation=90)
# ax.set_xticks([0, 50, 100])
# ax1.set_xticks([2017-01, 2018-01, 2019-01, 2020-01, 2021-01])


plt.legend(loc="upper right")


plt.xlabel('년도')
plt.ylabel('단위면적(㎡)당 금액(만원)')


ax2 = ax1.twinx()
plt.ylabel('출생자 수')

ax2.plot(g3, color='deeppink',label="출생자 수")

plt.title('강남구')

plt.legend(loc="upper left")

plt.show()
```
